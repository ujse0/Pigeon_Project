```python
import matplotlib.pyplot as plt
import numpy as np

models = ['Pretrained', 'Baseline (100ep)', 'Final (100ep)', 'Final (200ep)']
x = np.arange(len(models))

precision = [0.0540, 0.9854, 0.9852, 0.9855]
recall = [0.2368, 0.9693, 0.9674, 0.9648]
f1_score = [0.0879, 0.9773, 0.9762, 0.9750]
map5095 = [0.0276, 0.9570, 0.9532, 0.9524]

width = 0.2

fig, ax = plt.subplots(figsize=(12, 7))


rects1 = ax.bar(x - 1.5*width, precision, width, label='Precision', color='blue')
rects2 = ax.bar(x - 0.5*width, recall, width, label='Recall', color='green')
rects3 = ax.bar(x + 0.5*width, f1_score, width, label='F1 Score', color='purple')
rects4 = ax.bar(x + 1.5*width, map5095, width, label='mAP@0.5:0.95', color='red')


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.4f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3pt 위로
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)

for rects in [rects1, rects2, rects3, rects4]:
    autolabel(rects)

ax.set_ylabel('Score', fontsize=13)
ax.set_ylim(0, 1.05)
ax.set_xticks(x)
ax.set_xticklabels(models, fontsize=12, rotation=15)
ax.set_title('Model Performance Comparison (Bar Chart)', fontsize=16)


ax.grid(True, axis='y', linestyle='--', alpha=0.7)


ax.legend(loc='lower right', fontsize=11, frameon=True)

plt.tight_layout()
plt.show()
```
